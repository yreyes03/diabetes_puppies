### SVM Glucose + Age
Used glucose and age as predictors for diabetes outcome. Used 70% of the data for training and 30% for testing.

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
```

```python
#spliting testing and training data
dat_ga = dat[['Glucose', 'Age', 'Outcome']] # keeping only variable needed for this model
X = dat_ga[['Glucose', 'Age']]
y = dat_ga['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=341)
```

```python
svm = SVC(kernel='linear')  # C defaults to 1
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
```

### Plot for Glucose + Age

```python
# Scatter plot of the test data showing the decision boundary
plt.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=y_test, cmap='viridis')

# Create the decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)

# Plot the decision boundary
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

plt.title("Test Data with SVM Decision Boundary")
plt.xlabel("Glucose")
plt.ylabel("Age")
plt.show()
```

### SVM Glucose + Insulin
Used glucose and insulin as predictors for diabetes outcome. Used 70% of the data for training and 30% for testing.

```python
#spliting testing and training data
dat_gi = dat[['Glucose', 'Insulin', 'Outcome']] # keeping only variables needed for this model
X2 = dat_gi[['Glucose', 'Insulin']]
y2 = dat_gi['Outcome']
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size = .3, random_state=341)
```

```python
svm2 = SVC(kernel='linear') 
svm2.fit(X_train2, y_train2)
y_pred2 = svm2.predict(X_test2)
```

### Plot for Glucose + Insulin

```python
# Scatter plot of the test data showing the decision boundary
plt.scatter(X_test2.iloc[:, 0], X_test2.iloc[:, 1], c=y_test2, cmap='viridis')

# Create the decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)

# Plot the decision boundary
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

plt.title("Test Data with SVM Decision Boundary")
plt.xlabel("Glucose")
plt.ylabel("Insulin")
plt.show()
```
